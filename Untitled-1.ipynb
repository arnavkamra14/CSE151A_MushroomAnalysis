```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ast

# Took range of cap-diameter and took average and reinput into column.
# Probably can expand, just have 2 features for now

pdf3 = pdf2.copy()
print(pdf3['cap-diameter'].head())

def convert_range_to_average(range_str):
    try:
        range_list = ast.literal_eval(range_str)
        return sum(range_list) / len(range_list)
    except Exception as e:
        return np.nan 

# Apply the conversion to the 'cap-diameter' column
pdf3['cap-diameter'] = pdf3['cap-diameter'].apply(convert_range_to_average)

correlation_matrix = pdf3[['cap-diameter', 'class']]  # include other numeric features

print(pdf3.head())

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
